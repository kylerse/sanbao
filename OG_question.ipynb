**统计套利策略**

Mike最近在课上学习了统计套利交易的简单思想, 他想自己编写程序来对市场上的品种(股票、期货、外汇)进行交易. 在众多资产中, 他打算选取两个品种来配对交易(选定之后就无法更改). 试帮他找出选取哪两个品种可以获得最大的利润, 且计算该利润.

配对交易简单介绍: 后补, 如下图

![title](pair.png)

Mike的简单策略: 选定两个标的后, 按之前所有日期两者价格的平均差值作为他对两者价差的"合理估计"(expanding window), 之前所有日期两者价差序列的标准差作为对风险的"合理估计". 在新的每一天, 如果选定两个品种的价差偏离历史平均一个标准差, Mike预期会发生均值回归, 故会相应在两个品种分别买入和卖出一手(即一买一卖进行对冲, 买卖的方向为假设发生均值回归会盈利的方向). 两个标准差则两手, 以此类推. 且为了积累一定的历史参考, Mike在第三天才开始交易, 以前三天的平均值和方差作为初始值.

数据格式:  
第一行输入两个正整数 n m, 分别是总的品种数目以及总的天数, 接着输入m行, 每行n个正整数代表各品种的价格  
输出为,两个品种的序号(从小到大),以及最后的总收益.

### 例如:

输入:
4 8  
18 19 16 20  
14 15 18 12  
15 18 13 11  
17 18 10 16  
10 15 17 14  
11 11 13 17  
10 15 14 11  
14 16 17 15  

输出:
1 3 38

#### 具体过程
此时选取第1和第3个资产可以取得最大利润.  
资产A价格序列:  18 14 15 17 10 11 10 14  
资产B价格序列:  16 18 13 10 17 13 14 17   


差价定义为A-B, 则意味着向上偏离历史均值, A高估,应做空, B应做多. 反之亦然  
**第4天**  
-历史平均差价:0.00, 历史方差:2.83, 今日差价:7  
-A持有到明天单位收益=A(Day5)-A(Day4)=-7, 持有A手数:-2, 收益:14  
-B持到到明天单位收益=B(Day5)-B(Day4)=7, 持有B手数:2, 收益:14  
-总收益:28, 累积收益:28  
**第5天**

-B持到到明天单位收益=B(Day6)-B(Day5)=-4, 持有B手数:-2, 收益:8  
-总收益:10, 累积收益:38
